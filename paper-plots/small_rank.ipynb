```python
with open('raw_tex.txt', 'r') as f:
    raw_tex = f.readlines()
```

```python
model_name_list = []
result_list = [[] for _ in range(9)]

for i in range(20, 37):
    text_list =  raw_tex[i].split('&')
    if not text_list[0].strip().startswith('%'):
        model_name_list.append(text_list[0].strip())
        for i in range(1, 10):
            res = re.findall(r'\d+\.*\d*', text_list[i])
            if len(res) == 2:
                result_list[i-1].append(float(res[0]))
            else:
                result_list[i-1].append(0)
```

```python
model_name_list
```

```python
result_list
```

```python
result_dict = {
    'model': model_name_list,
    'texas': result_list[0],
    'wisconsin': result_list[1],
    'cornell': result_list[2],
    'actor': result_list[3],
    'squirrel': result_list[4],
    'chameleon': result_list[5],
    'cora': result_list[6],
    'citeseer': result_list[7],
    'pubmed': result_list[8]
}
```

```python
df = pd.DataFrame.from_dict(result_dict)
df
```

```python
datasets = df.columns.tolist()[1: ]
datasets
```

```python
for i in range(len(datasets)):
    column_name = datasets[i] + '_rank'
    df[column_name] = df[datasets[i]].rank(ascending=False)
```

```python
df['average_rank'] = 0
for i in range(len(datasets)):
    column_name = datasets[i] + '_rank'
    df['average_rank'] += df[column_name]

df['average_rank'] /= 9
```
